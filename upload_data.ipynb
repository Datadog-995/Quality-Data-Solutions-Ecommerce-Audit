```python
import pandas as pd

# 1. Load your live ecommerce dataset from your environment
csv_path = 'Ecommerce-Order-Data-Cleaning-Sample-3-copy-xlsx.csv'
df_ecommerce = pd.read_csv(csv_path)

# 2. Extract numbers, handle negatives, and strip out everything else
def clear_non_numbers(val):
    val_str = str(val).strip()
    try:
        # Convert to float to verify it's a valid number
        num = float(val_str)
        # Keep the exact number formatting (including negatives if applicable)
        return f"{num:.2f}"
    except ValueError:
        # If it's a dash '-', letters, or blank, return a completely empty string
        return ""

df_ecommerce['Price'] = df_ecommerce['Price'].map(clear_non_numbers)

# 3. Update the Price_Audit_Flag to 'Unknown' for these empty cells
df_ecommerce['Price_Audit_Flag'] = df_ecommerce['Price'].apply(lambda x: 'Unknown' if x == "" else 'Valid')

# 4. Save the full updated dataset back to your file
df_ecommerce.to_csv(csv_path, index=False)

# 5. Preview to verify that cells without numbers are completely blank boxes
df_ecommerce[['Order ID', 'Product Name', 'Price', 'Price_Audit_Flag']].head(10)
```
